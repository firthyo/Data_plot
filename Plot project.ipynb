```python
BldgType=Housing["BldgType"].unique()
Count=Housing["BldgType"].value_counts()
x = BldgType
plt.bar(x,Count)
plt.ylabel("BldgType")
plt.title("Housing Type")
plt.show()
```

```python
filtered_df = Housing[Housing['LotArea'] <= 20000]
```

```python
x = filtered_df["LotArea"].values
y = filtered_df["SalePrice"].values
plt.scatter(x,y)
plt.xlabel("LotArea")
plt.ylabel("SalePrice")
plt.show()
```

```python
x = filtered_df["LotArea"].values
y = filtered_df["SalePrice"].values
m,b = np.polyfit(x,y,1)
line_y = m*x+b
plt.plot(x,line_y,linewidth=3,color="red",alpha=0.8,linestyle="-",label='BFit')
plt.scatter(x,y,label= 'SalePrice')
plt.xlabel("LotArea")
plt.ylabel("SalePrice")
plt.title("Housing")
plt.legend()
plt.show()
```
