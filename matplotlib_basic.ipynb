Plot an exponential curve representing y=x**2 with the x-axis ranging from 0-100.
Adjust the curve so that it is a green dotted line.
Add a legend and call it exponential.

Hint: Use np.arange(n) to create a numpyarray from 0-100
Hint: Numpy makes it easy to multiply arrays! You can square each number in the X arrayby using the exponent
operator against the array like so: X**2

```python
n = 100
np.arange(n)
# x= linspace 0-100 
x = np.linspace(0,n)
y = x**2

plt.title("exponential graph")
plt.plot(x,y,linewidth=3,color="Green",alpha=1,linestyle=":",label="exponential")
plt.legend()
```
